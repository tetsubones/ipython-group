```python
import pandas as pd

sales = pd.read_clipboard()

sales
```

```python
employer = pd.read_clipboard()

employer
```

```python
#従業員IDでの売り上げ合計のDataframeを作成
emptotal = sales.groupby(['emp_id'])['total']
d = pd.DataFrame(emptotal.sum())

#mergeにindexは使えないのでcolumnに入れる
d['emp_id'] = d.index

#確認用出力
employer[['emp_id','emp_name']]
```

```python
#従業員IDをキーにしてデータを結合
d = pd.merge(d,employer[['emp_id','emp_name']],on='emp_id')
d
```
